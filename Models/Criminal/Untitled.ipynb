```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('criminal_train.csv')
df_test=pd.read_csv('criminal_test.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop(['PERID','Criminal'],axis=1)
```

```python
scale=StandardScaler()
X=scale.fit_transform(X)
```

```python
X_test=df_test.drop(['PERID'],axis=1)
```

```python
X_test.head()
```

```python
X_test=scale.transform(X_test)
```

```python
X_test.shape
```

```python
y=df['Criminal']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
```

```python
logis=LogisticRegression()
```

```python
logis.fit(X,y)
```

```python
logis.score(X,y)
```

```python
preds=logis.predict(X_test)
```

```python
subs=pd.DataFrame({'PERID':df_test['PERID'],'Criminal':preds})
subs.to_csv('logistic.csv',index=False,header=True,columns=['PERID','Criminal'])
```

```python
subs.head()
```
